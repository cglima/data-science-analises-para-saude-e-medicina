```python
import pandas as pd

confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
confirmed.head()
```

```python
# saber o formato da tabela 
confirmed.shape
```

```python
# selecionar somente uma coluna
#serie de valores 
confirmed["Country/Region"]
```

```python
#buscar valores únicos 
confirmed["Country/Region"].unique()
```

```python
confirmed["Province/State"].unique()
```

```python
#pegar uma amostra aleatória dentro do dataframe 
confirmed.sample(20)
```

```python
# pegar uma linha especifica - com índice 0
#localiza pela linha
confirmed.iloc[0]
```

```python
#procurar pelo nome
# loc localiza pelo índice
confirmed.set_index("Province/State").loc["Anhui"]
```

```python
confirmed.set_index("Province/State").iloc[1]
```

```python
# contar qtas vezes aparece o "país"
# contar quantas ocorrências aparecem

#valor falso pois o erro é muito grande - subnotificações
# não vou considerar esta aalise pois tem muta campo vazio (mta disparidade no padrão)
confirmed["Country/Region"].value_counts()
```

```python
# agrupamento e soma
confirmed_by_country = confirmed.groupby("Country/Region").sum()
confirmed_by_country
```

```python
# 10 aleatórios
confirmed_by_country.sample(10)
```

```python
# casos confirmados na China por dia
confirmed_by_country.loc["China"][2:]
```

```python
#plotar o gráfico
import matplotlib.pyplot as plt

# gráfico extremamente apertado
# estamos sendo longe do adequado ao passar uma mensagem errada

plt.figure(figsize=(2,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.show()
```

```python
# por isso em data visualization é FUNDAMENTAL
# estarmos atentos aos DOIS EIXOS
# esse aqui está tentando comparar o crescimento do momento inicial ao momento atual

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.show()
```

```python
# exemplo ruim de manusear o eixo y

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.ylim(40000, 80000)
plt.show()
```

```python
# exemplo ruim de manusear o eixo y
# aqui a mensagem ficou novamente diferente

# esse aqui está tentando comparar o momento inical com 300mil

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.ylim(0, 300000)
plt.show()
```

```python
# exemplo ruim de manusear o eixo y
# aqui a mensagem ficou novamente diferente

# esse aqui está tentando comparar o momento inical com 1.386bi (população da china em 2017)
# aqui a mensagem é novamente completamente diferente
# e pode ser uma mensagem ruim, até mesmo errada. Poderia ser uma doença grave
# mas ao usarmos o eixo y com a população geral, fingimos que ela não é grave
# sendo que na realidade esse gráfico não POSSUI INFORMAÇÃO NENHUMA sobre a gravidade
# da doença

# pq? só tem caso confirmado, não fala em letalidade nem em incidência
# quais são as províncias infectadas? são de toda a China? e estou comparando
# com a população de toda a China. Está errado.

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Gráfico errado. Passa uma mensagem errada, com comparações injustas")
plt.ylim(0, 1386000000)
plt.show()
```
