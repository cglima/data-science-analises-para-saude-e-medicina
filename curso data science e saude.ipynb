```python
# contar qtas vezes aparece o "paÃ­s"
# contar quantas ocorrÃªncias aparecem

#valor falso pois o erro Ã© muito grande - subnotificaÃ§Ãµes
# nÃ£o vou considerar esta aalise pois tem muta campo vazio (mta disparidade no padrÃ£o)
confirmed["Country/Region"].value_counts()
```

```python
# agrupamento e soma
confirmed_by_country = confirmed.groupby("Country/Region").sum()
confirmed_by_country
```

```python
# 10 aleatÃ³rios
confirmed_by_country.sample(10)
```

```python
# casos confirmados na China por dia
confirmed_by_country.loc["China"][2:]
```

```python
#plotar o grÃ¡fico
import matplotlib.pyplot as plt

# grÃ¡fico extremamente apertado
# estamos sendo longe do adequado ao passar uma mensagem errada

plt.figure(figsize=(2,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.show()
```

```python
# por isso em data visualization Ã© FUNDAMENTAL
# estarmos atentos aos DOIS EIXOS
# esse aqui estÃ¡ tentando comparar o crescimento do momento inicial ao momento atual

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.show()
```

```python
# exemplo ruim de manusear o eixo y

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.ylim(40000, 80000)
plt.show()
```

```python
# exemplo ruim de manusear o eixo y
# aqui a mensagem ficou novamente diferente

# esse aqui estÃ¡ tentando comparar o momento inical com 300mil

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("Casos confirmados em China")
plt.ylim(0, 300000)
plt.show()
```

```python
# exemplo ruim de manusear o eixo y
# aqui a mensagem ficou novamente diferente

# esse aqui estÃ¡ tentando comparar o momento inical com 1.386bi (populaÃ§Ã£o da china em 2017)
# aqui a mensagem Ã© novamente completamente diferente
# e pode ser uma mensagem ruim, atÃ© mesmo errada. Poderia ser uma doenÃ§a grave
# mas ao usarmos o eixo y com a populaÃ§Ã£o geral, fingimos que ela nÃ£o Ã© grave
# sendo que na realidade esse grÃ¡fico nÃ£o POSSUI INFORMAÃ‡ÃƒO NENHUMA sobre a gravidade
# da doenÃ§a

# pq? sÃ³ tem caso confirmado, nÃ£o fala em letalidade nem em incidÃªncia
# quais sÃ£o as provÃ­ncias infectadas? sÃ£o de toda a China? e estou comparando
# com a populaÃ§Ã£o de toda a China. EstÃ¡ errado.

plt.figure(figsize=(10,5))
confirmed_by_country.loc["China"][2:].plot()
plt.title("GrÃ¡fico errado. Passa uma mensagem errada, com comparaÃ§Ãµes injustas")
plt.ylim(0, 1386000000)
plt.show()
```
